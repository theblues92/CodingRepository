# Crawling 네이버 뉴스

### 얻고자 하는 정보는 네이버 기준 많이 본 뉴스 순위, 제목, 미디어

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import csv
```

```python
html = urlopen("https://news.naver.com/main/ranking/popularDay.nhn?rankingType=popular_day&sectionId=105&date=20190322")
```

```python
Object = BeautifulSoup(html, "html.parser")
Object
```

```python
for n in range(0,30):
    crt = Object.find("li", {"class":"ranking_item is_num"+str(n+1)})
    title = crt.select('a')[-1].get_text()
    media = crt.select('div')[-2].get_text()
    print(n+1, title, media)
```

```python
with open('Crawling_Drill.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['rank', 'title', 'media'])

    for n in range(0,30):
        crt = Object.find("li", {"class":"ranking_item is_num"+str(n+1)})
        title = crt.select('a')[-1].get_text()
        media = crt.select('div')[-2].get_text()
        writer.writerow([n+1, title, media])
```
