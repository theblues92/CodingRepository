```python
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = tensorflow.keras.datasets.boston_housing.load_data(path='boston_housing.npz')
```

```python
model = Sequential()
model.add(Dense(1, input_shape=[13]))
model.compile(optimizer='rmsprop', loss='mse')
model.summary()
```

```python
model.fit(x_train, y_train, epochs=1000)
```

```python
model.evaluate(x_test, y_test)
```

하나는 x, y train 으로 불러와서 모델을 만들어서 훈련시켰고, x, y test 는 그것을 다시 evaluate하기 위해서.  
데이터의 약 80%는 훈련할 때 쓰고(fit)  
그리고 훈련할 때 안 쓴 데이터로 검증을 해야 함 약 20%(evaluate)  
fit은 train이고 evaluate는 test임.  
predict는 실제 상품화 시킬 때.
