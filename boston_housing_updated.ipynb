```python
# description
# https://www.kaggle.com/c/boston-housing

# TensorFlow and tf.keras
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation

# Helper libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = tensorflow.keras.datasets.boston_housing.load_data(path='boston_housing.npz')
```

```python
print(x_train.shape); print(y_train.shape)
```

```python
print(x_test.shape); print(y_test.shape)
```

x_train 의 shape 이 (404, 13) 이었으니 plot을 하려면 13 개 중 한 개를 선택해야 함.  
x_train\[:,1\]이 첫 번째 vector를 plot 한 것임.

```python
plt.plot(x_train[:,0], y_train, '.')
```

y축은 집값, x축은 crime rate  
만약 대충 우리 멋대로 plot을 그려보면 linearity가 아니라 오목한 포물선 느낌으로 볼 수 있음.  
즉 linearity로만 인공신경망 훈련하면 잘 안 나오겠구나를 알 수 있음, linear한 관계라고 보기는 좀 어렵겠구나 라고 느껴야 함.

```python
plt.plot(x_train[:,1], y_train, '.')
```

```python
model = Sequential()
model.add(Dense(1, input_shape=[13])) # input - 13 / output - 1
model.compile(optimizer='rmsprop', loss='mse') # regression의 경우는 이런 식으로 적음
model.summary()
```

summary에서 알아야 할 건 parameter에 14라고 되어 있음.  
우리가 알아야 할 것은 13 아닌가라고 생각할 수 있음.  
y 절편 값을 정해주는 b값을 생각해야 함.

```python
model.fit(x_train, y_train, epochs=1000)
```

산점도 생각해 봐라. 12차원(?)과 실제 target 값과의 차이가 loss.  
epoch가 늘어날수록 loss가 점점 줄어듦.  
훈련이 되고 있다는 것임

이제 hidden layer를 넣어보자.

```python
model = Sequential()
model.add(Dense(10, input_shape=[13]))
model.add(Dense(20))
model.add(Dense(1))
model.compile(optimizer='rmsprop', loss='mse')
model.summary()
```

hidden layer 두 개 들어갔음

잘 봐야 할 것이 있음.  
우선 node를 생각해보자, 첫 번째 layer에 13개(사실 14개),  
hidden layer 첫 번째에는 10개(사실 11개),  
그다음 hidden layer에는 20개(사실 21개),  
그 다음 마지막 layer에는 1개이다.  
  
parameter를 잘 생각해보면 14\*10이어야 한다.  
그러니까 parameter를 그릴 때 시작점은 'b'(1 node) 값을 포함해야 하고,  
parameter를 그릴 때 종착점은 'b'(1 node) 값을 포함하면 안 된다.

```python
model.fit(x_train, y_train, epochs=1000)
```

여기서 layer 들 다 linear 한 거라 hidden layer 많이 넣어도 loss값 걍 비슷함.  
따라서 not linearity 넣어야 함.

```python
model = Sequential()
model.add(Dense(10, input_shape=[13], activation = 'relu'))
model.add(Dense(20))
model.add(Dense(1))
model.compile(optimizer='rmsprop', loss='mse')
model.summary()
```

```python
model.evaluate(x_test, y_test)
```

dataset 을 불러 그것을 할당할 때를 살펴보면  
한 쌍은 x_train, y_train 으로,  
또 한 쌍은 x_test, y_test 로 받았다.  

데이터의 약 80%는 훈련할 때 쓰고(fit)  
그리고 검증은 훈련할 때 쓰지 않은 데이터를 이용한다, 즉 나머지 약 20%(evaluate)  
  
결국 우리가 할당한 그 변수쌍을 보면 fit 같은 경우는 train, evaluate는 test이다.

```python
print(model.input_shape)
print(model.output_shape)
```

Boston_Housing dataset에서 input이 13개 였고 target이 1개 였으므로,  
위에처럼 13, 1로 나온다.

predict는 실제 상품화 시킬 때의 것으로,  
우리가 얻어낸 그 행렬값, parameter를 들고,  
우리가 input을 탁 집어넣어 탁 얻는 예측값인 것이다

```python
# INPUT::
# 1. crim: per capita crime rate by town.
# 2. zn: proportion of residential land zoned for lots over 25,000 sq.ft.
# 3. indus: proportion of non-retail business acres per town.
# 4. chas: Charles River dummy variable (= 1 if tract bounds river; 0 otherwise).
# 5. nox: nitrogen oxides concentration (parts per 10 million).
# 6. rm: average number of rooms per dwelling.
# 7. age: proportion of owner-occupied units built prior to 1940.
# 8. dis: weighted mean of distances to five Boston employment centres.
# 9. rad: index of accessibility to radial highways.
# 10. tax: full-value property-tax rate per $10,000.
# 11. ptratio: pupil-teacher ratio by town.
# 12. black: 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town.
# 13. lstat: lower status of the population (percent).

# TARGET::
# medv: median value of owner-occupied homes in $1000s.

model.predict(np.array([0, 1, 50, 1, 0, 5, .5, 10, 1000, 100, 10, 100, 1]).reshape(1,13))
```

```python
boston_housing = ['crim', 'zn', 'indus', 'chas', ' nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'black', 'lstat']

for i in range(13):
    plt.subplot(3,5,i+1) # 이 부분 조금 더 알아볼 것
    plt.plot(x_train[:,i], y_train, '.')
    plt.title(boston_housing[i])
plt.show()
```
