re package의 search 함수의 입력은 두 개임  
첫 번째 input은 spot할 것,  
두 번째 input은 spot을 어디에서 하느냐,  
output은 두 번째 input에서 첫 번쨰 input이 있는 범위를 보여주고,  
match 되는 string을 보여준다

```python
result = [w for w in wordlist if re.search('ed$', w)][:10]
print(result)
```

wordlist 안에서 w로 받아 for loop을 돌리는데,  
만약 그것의 범위 안에 ed로 끝나는 것이 있다면,  
즉 그 단어가 ed로 끝난다면 list의 형태로 w를 쌓고,  
그것의 10개까지의 단어를,  
result에 넣어,  
그것을 print해라

```python
result = [w for w in wordlist if re.search('ed$', w)]
print(result[:10])
```

위의 cell도 그 위의 cell과 거의 똑같긴 한데,  
다만 전자가 result 안에 10개의 단어까지만 들어갔다면,  
후자는 result 안에 ed로 끝나는 단어를 다 집어넣고,  
그것을 10개까지 출력한 값이 나오는 것이다

^은 시작, .은 하나의 character.

```python
result = [w for w in wordlist if re.search('^..j..t..$', w)]
print(result[:10])
```

^는 string의 시작을 의미하고,  
$는 string의 끝을 의미하는데,  
.은 하나의 무작위 character를 의미하므로,  
search spot 하는 단어는 8개의 character를 가지고,  
세 번째는 j character가,  
여섯 번째는 t character가 나오는 것이다

따라서 시작과 끝이 정해져 있으므로 8개의 character  
그리고 세 번째는 j 여섯 번째는 t가 나옴

```python
result = [w for w in wordlist if re.search('^[ghi][mno][jlk][def]$', w)]
print(result[:10])
```

bracket안에 나오는 character들은 그것들 중 임의의 하나라는 뜻이다

```python
result = [w for w in wordlist if re.search('^[ah]+$', w)]
print(result[:10])
```

\+ sign은 바로 전의 것이 한 개 이상 나올 수 있다는 뜻이다  
즉, \[ah\]\[ah\]\[ah\]\[ah\]...\[ah\] 와 같은 식의 것으로 파악할 수 있다

```python
result = [w for w in wordlist if re.search('^[ah]+', w)]
print(result[:10])
```

위의 cell과는 다른 점은 \[ah\]\[ah\]\[ah\]\[ah\]... 이라는 것이다  
즉, 시작만 a or h로 몇 번 반복되어도 상관없고 그 뒤에는 어떤 character가 와도 된다는 것이다

```python
wsj = sorted(set(nltk.corpus.treebank.words()))
wsj
```

해당 위의 line 역시 말뭉치를 가져온 건데, word street journal에 나왔던 단어이다

위에 cell 두 개를 통해 알 수 있는 건,  
소문자 a-z와 대문자 A-Z가 구별된다는 점이고,  
소문자 a-z에는 특수기호까지 들어간다는 것이다

```python
wordlist = [w for w in wsj if re.search('^[0-9]{4}$', w)]
print(wordlist[:10])
```

curly bracket은 그 앞의 것을 몇 번 반복하냐를 나타내는 것이다

```python
wordlist = [w for w in wsj if re.search('^[0-9][0-9]{3}$', w)]
print(wordlist[:10])
```

따라서 위의 cell은 그 위의 cell과 동일한 결과를 보여준다

```python
wordlist = [w for w in wsj if re.search('^[0-9]+-[a-z]{3,5}$', w)]
print(wordlist[:10])
```

\- 는 그냥 대시기호이고,  
a to z가 3번에서 5번 반복되면서 끝나야 한다

```python
wordlist = [w for w in wsj if re.search('^[a-z]{5,}-[a-z]{2,3}-[a-z]{,6}$', w)]
print(wordlist[:10])
```

a to z가 5번 이상 나오며 시작되고,  
\- 기호가 나오고,  
a to z가 2번에서 3번 나오고,
다시 - 기호가 나오고,  
다시 a to z가 6번이하로 반복되며 끝나는 것
  
근데 6번 이하라는 거는 0번도 포함하는 걸까?

```python
wordlist = [w for w in wsj if re.search('^[0-9]{,1}', w)]
print(wordlist[:10])

wordlist = [w for w in wsj if re.search('^[0-9]', w)]
print(wordlist[:10])
```

```python
wsj = sorted(set(nltk.corpus.treebank.words()))
fdist = nltk.FreqDist(vs for word in wsj for vs in re.findall(r'[aeiou]{2,}', word))
fdist.most_common(10)
```

모음이 두 개 이상의 character가 들어있기만 하면,  
그것을 찾아와서 list로 만들었음.

```python
regexp = r'^[AEIOUaeiou]+|[AEIOUaeiou]+$|[^AEIOUaeiou]'
def compress(word):
    pieces = re.findall(regexp, word)
    return ''.join(pieces)

english_udhr = nltk.corpus.udhr.words('English-Latin1')
print(nltk.tokenwrap(compress(w) for w in english_udhr[:75]))
```

```python
rotokas_words = nltk.corpus.toolbox.words('rotokas.dic')
cvs = [cv for w in rotokas_words for cv in re.findall(r'[ptksvr][aeiou]', w)]
cfd = nltk.ConditionalFreqDist(cvs) # CFD works pairwise only
cfd.tabulate()
```
